```python
import re
import copy

def Apriori_Algorithm(Args):
    MinSupport=float(input("Enter Minimum Support Value (in %):- "))
    MinConfidence=float(input("Enter Minimum Confidence Value (in %):- "))

    # reading the file and storing Transactionactions in list
    with open(Args,'r') as fileData:
            TransactionList=fileData.readlines()
    NoOfTransactions=len(TransactionList)
    Transaction=[]
    C={}
    suppDict={}
    Unit=[]
    Data=[]

    for line in TransactionList:
        RefinedLine=re.sub(r'T\d{1,5}',"",line).strip()
        ItemsOfLine=RefinedLine.split(',')
        Transaction.append(ItemsOfLine)
        for Index in ItemsOfLine:
            if (tuple([Index])) in C:
                C[tuple([Index])]+=1
            else:
                C[tuple([Index])]=1

    # checking the minimum support value of the items
    Items=[]
    for Index in C:
        if (C[Index]/NoOfTransactions)*100 >= MinSupport:
            Items.append(Index)
    
    # choosing the items with minimum support
    for Index in Items:
        suppDict[Index]=C[Index]
    Data.append(suppDict)

    for Index in Data[0]:
        Unit.append(Index)

    t=0
    while(1):
        if Data[t]:
            Data.append({})
            for ElementGroup in Data[t]:
                for Index in Unit:

                    if not set(list(Index))&set(list(ElementGroup)):
                        TmpGroup=ElementGroup+Index

                        if TmpGroup not in Data[t+1]:
                            Data[t+1][TmpGroup]=0
            TmpGroupList=[]
            TmpList=[]
            for Index in Data[t+1]:  # Here we delete the duplicate Transactionactions eg: 1,2,3 = 2,3,1
                if set(Index) in TmpGroupList and Index not in TmpList:
                    TmpList.append(Index)
                elif set(Index) not in TmpGroupList:
                    TmpGroupList.append(set(Index))

            for Index in TmpList:
                del Data[t+1][Index]

            t+=1
            for TansLine in Transaction:
                for LKey in Data[t]:
                    if set(list(LKey))<=set(TansLine):
                        Data[t][LKey]+=1

            Items2=[]
            for Index in Data[t]:
                if (Data[t][Index]/NoOfTransactions)*100<MinSupport:
                    Items2.append(Index)
            for Index in Items2:
                del Data[t][Index]

        else:
            break
    del Data[t]

    print ('\nFrequent itemsets with their support value:-\n')
    for Index1 in Data:
        for Index2 in Index1:
            print (Index2,':',Index1[Index2])
    print ('\n')

    TmpList=[]
    LL=[]
    Temporary_Set_List=[]
    for Index1 in Data:
        for Index2 in Index1:
            for Index3 in Index2:
                if (set(list(Index2))-set(list(tuple([Index3])))):
                    LL.append([[Index3],list(set(list(Index2))-set(list(tuple([Index3]))))])
                    Temporary_Set_List.append(set(Index3))

    for Index1 in LL:
        if len(Index1[1])>1:
            for Index2 in Index1[1]:
                tmpKey=copy.deepcopy(Index1)
                tmpKey[0].append(Index2)
                tmpKey[1].remove(Index2)
                if set(tmpKey[0]) not in Temporary_Set_List:
                    LL.append(tmpKey)
                    Temporary_Set_List.append(set(tmpKey[0]))

    print('The association rules with their Support and Confidence are as follows :-\n')
    for Index1 in LL:
        for Index2 in Data:
            for Index3 in Index2:
                if set(list(Index3))==set(Index1[0]):
                    FirstInt=Index2[Index3]
                if set(list(Index3))==set(Index1[0]+Index1[1]):
                    TotalInt=Index2[Index3]

        if (TotalInt/FirstInt)*100>=MinConfidence:
            print (Index1[0],'->',Index1[1],'  Support:-',(TotalInt/NoOfTransactions)*100,'Confidence:-',(TotalInt/FirstInt)*100,')')


while(1):
    print('\n')
    print('*****************************Welcome********************************')
    print ('Enter 1 for BJ\'s')
    print ('Enter 2 for Walmart')
    print ('Enter 3 for Kmart')
    print ('Enter 4 for Amazon')
    print ('Enter 5 for Costco')
    Input_file=input('Please select a Transaction file (enter q to exit):')
    if Input_file=='1':
        print ("\nYou have selected BJ\'s\n")
        Apriori_Algorithm('Transactions_BJ.txt')
    elif Input_file=='2':
        print ("\nYou have selected Walmart\n")
        Apriori_Algorithm('Transactions_Walmart.txt')
    elif Input_file=='3':
        print ("\nYou have selected Kmart\n")
        Apriori_Algorithm('Transactions_Kmart.txt')
    elif Input_file=='4':
        print ("\nYou have selected Amazon\n")
        Apriori_Algorithm('Transactions_Amazon.txt')
    elif Input_file=='5':
        print ("\nYou have selected Costco\n")
        Apriori_Algorithm('Transactions_Costco.txt')
    elif Input_file=='q':
        break
    else:
        print('Please enter again.')
```
